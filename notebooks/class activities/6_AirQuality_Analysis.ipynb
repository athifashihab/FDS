**Step-by-Step Instructions**
- **Step 1**: Introduction to the Dataset
    - Goal : Familiarize yourself with the dataset and its columns.
      - **Task 1.1**: Load the dataset into a `table` and display the first few rows.

```python
# Task 1.1: Load the dataset into a Table and display the first few rows
path_data = ''
airquality = Table.read_table(path_data + 'California_airquality.csv')

airquality.show(5)
```

**Step 2**: Sorting Rows
  - Goal : Teach students how to sort rows based on specific criteria.
    - **Task 2.1** : Sort the dataset by Daily Mean PM2.5 Concentration in descending order to find the days with the worst air quality.

```python
# Task 2.1: Sort by Daily Mean PM2.5 Concentration in descending order
sorted_by_pm25 = airquality.sort('Daily Mean PM2.5 Concentration', descending=True)
sorted_by_pm25.show(10)
```

```python
# To understand the sorting more clearly
sorted_by_pm25.select('Date', 'Daily Mean PM2.5 Concentration', 'CBSA_NAME', 'COUNTY').show(10)
```

**Teams should discuss why certain days might have had poor air quality (e.g., wildfires)!**

> **Discussion:** The days with the highest PM2.5 concentrations fall in August–September 2020. This aligns with California's record-breaking 2020 wildfire season. Wildfires release massive amounts of fine particulate matter (PM2.5) into the air, dramatically worsening air quality across wide regions. Counties like Santa Cruz, San Luis Obispo, Plumas, and Mariposa were particularly hard hit.


- **Task 2.2** : Sort the dataset alphabetically by `CBSA_NAME` to group data by region.

```python
# Task 2.2: Sort alphabetically by CBSA_NAME
sorted_by_cbsa = airquality.sort('CBSA_NAME')
sorted_by_cbsa.show(10)
```

```python
# To understand the sorting more clearly
sorted_by_cbsa.select('Date', 'CBSA_NAME', 'Daily Mean PM2.5 Concentration').show(10)
```

**Step 3**: Selecting Rows
  - Goal : Practice filtering rows based on specific conditions.
      - **Task 3.1** : Extract rows where the Daily Mean PM2.5 Concentration exceeds 50 µg/m³.

```python
# Task 3.1: Filter rows where PM2.5 > 50 µg/m³
high_pm25 = airquality.where('Daily Mean PM2.5 Concentration', are.above(50))
print(f'Number of days with PM2.5 > 50 µg/m³: {high_pm25.num_rows}')
high_pm25.show(5)
```

```python
# To understand the selection more clearly
high_pm25.select('Date', 'Daily Mean PM2.5 Concentration', 'CBSA_NAME', 'COUNTY').show(5)
```

**Teams should count how many such days occurred and discuss the health implications.**

> **Discussion:** A total of **354 days** recorded PM2.5 concentrations above 50 µg/m³ across the dataset. PM2.5 > 35 µg/m³ is "Unhealthy" for all people per EPA guidelines. At these levels, everyone may experience health effects: respiratory irritation, reduced lung function, and increased risk of cardiovascular events. Sensitive groups (children, elderly, people with lung/heart conditions) are especially at risk. Most of these extreme readings occurred during the August–September 2020 wildfire events.


- **Task 3.2** : Find all rows corresponding to `Fresno, CA`.

```python
# Task 3.2: Find all rows for Fresno, CA
fresno = airquality.where('CBSA_NAME', are.equal_to('Fresno, CA'))
print(f'Number of rows for Fresno, CA: {fresno.num_rows}')
fresno.show(5)
```

```python
# To understand more clearly
fresno.select('Date', 'Daily Mean PM2.5 Concentration', 'CBSA_NAME').show(5)
```

**Teams should calculate the average PM2.5 concentration for Fresno during this period.**

```python
# Calculate average PM2.5 for Fresno
fresno_avg = np.mean(fresno.column('Daily Mean PM2.5 Concentration'))
print(f'Average PM2.5 concentration for Fresno, CA: {fresno_avg:.2f} µg/m³')
```

An average PM2.5 concentration of 11.60 µg/m³ in Fresno falls within the “Good” air quality range (below 12 µg/m³).Even though the average is “Good,” it may still hide short periods of higher pollution, so residents could still experience temporary unhealthy air days.

**Step 4**: Creating New Columns
- Goal : Teach students how to compute new metrics and add them to the table
    - **Task 4.1** : Add a column indicating whether the air quality is `"Good" (PM2.5 < 12)`, `"Moderate" (12 ≤ PM2.5 < 35)`, or `"Unhealthy" (PM2.5 ≥ 35)`.

```python
# Task 4.1: Add 'Air Quality Category' column
def classify_air_quality(pm25):
    if pm25 < 12:
        return 'Good'
    elif pm25 < 35:
        return 'Moderate'
    else:
        return 'Unhealthy'

categories = airquality.apply(classify_air_quality, 'Daily Mean PM2.5 Concentration')
airquality_with_cat = airquality.with_column('Air Quality Category', categories)
airquality_with_cat.select('Date', 'Daily Mean PM2.5 Concentration', 'Air Quality Category').show(5)
```

**Task 4.2** : Calculate the percentage of days in each category for Fresno.

```python
# Task 4.2: Calculate percentage of days in each category for Fresno
fresno_cat = airquality_with_cat.where('CBSA_NAME', are.equal_to('Fresno, CA'))
total_fresno = fresno_cat.num_rows

good_count      = fresno_cat.where('Air Quality Category', are.equal_to('Good')).num_rows
moderate_count  = fresno_cat.where('Air Quality Category', are.equal_to('Moderate')).num_rows
unhealthy_count = fresno_cat.where('Air Quality Category', are.equal_to('Unhealthy')).num_rows

print('Air Quality Category percentages for Fresno, CA:\n')
print(f'Good:       {good_count / total_fresno * 100:.2f}%')
print(f'Moderate:   {moderate_count / total_fresno * 100:.2f}%')
print(f'Unhealthy:   {unhealthy_count / total_fresno * 100:.2f}%')
```

**Step 5**: Group Challenge
- Goal :  solve a complex problem.
    - **Challenge** : Identify the top 3 regions (CBSA Names) with the highest average PM2.5 concentrations

```python
# Challenge: Identify the top 3 regions (CBSA Names) with the highest average PM2.5

top3_regions = (
    airquality
    .select('CBSA_NAME', 'Daily Mean PM2.5 Concentration')  # keep only needed columns
    .group('CBSA_NAME', np.mean)                            # group by region, compute mean
    .sort('Daily Mean PM2.5 Concentration mean', descending=True)  # highest first
    .take(np.arange(3))                                     # take top 3 rows
)

top3_regions.show()
```
